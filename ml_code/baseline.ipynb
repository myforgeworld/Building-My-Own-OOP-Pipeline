# Baseline-модель

В этой части работы я создам *baseline-модель*, чтобы определить, какие изменения помогут улучшить качество модели. В качестве baseline я буду использовать базовую модель **LinearRegression()**. Этот подход основан на совете **Andrew Ng** — сначала сделать проект в стиле **“fast and dirty”**, то есть быстро получить первый рабочий результат, а затем постепенно улучшать его. Чтобы понять улучшение модели я выбрал метрику **mean_squared_error()**.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Загружаем данные и делаем shuffle
df = pd.read_csv(r"..\data\preprocessed\final_data.csv", index_col=0)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```

```python
# train test split
target = 'realSum'

y = df[target]
X = df.drop(columns=target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# baseline
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print(f"Mean squere error: {mean_squared_error(y_test, y_pred)}")
```
